<center><h1>Handling Missing Values</h1></center>

na_values - set NaN manually  <br>
keep_default_na - Don't set any value to NaN  <br> 
na_filter - Used to find missing values & interpret them <br>

```python
import pandas as pd
```

```python
f = pd.read_csv('F:\\dataTut24(2).csv')                                                   # Read CSV
f
```

Pandas has fn, If there is missing data or another str(#N/A , nan) , then it set all to NaN. <br>
But if there is also unnecessary data, so we can manually set there to NaN using **na_values**

```python
f = pd.read_csv('F:\\dataTut24(3).csv')                                                   # Read CSV
f
```

# Set unnecesary data place to NaN ex- not available

```python
f = pd.read_csv('F:\\dataTut24(3).csv' , na_values='not availabe')                                                # Read CSV
f
```

# Consider more than 1 values to NaN use na_values with list

```python
f = pd.read_csv('F:\\dataTut24.csv')                                                # Read CSV
f
```

```python
f = pd.read_csv('F:\\dataTut24.csv' ,na_values=['not availabe' , 'no val'])                         # Read CSV
f
```

# Set more than 1 col values to NaN By specify column name, use na_values with dict

```python
f = pd.read_csv('F:\\dataTut24.csv')                                                    # Read CSV
f
```

```python
f = pd.read_csv('F:\\dataTut24.csv' ,na_values={'Companies' :['not availabe' , 'no val'] , 'Profit' : 'no val'})
f
```

<h3>Default Pandas set NaN , If you want pandas don't set NaN Use Keep_default_na fn</h3

```python
f = pd.read_csv('F:\\dataTut24(4).csv')
f
```

```python
f = pd.read_csv('F:\\dataTut24(4).csv' , keep_default_na=False)                         # By default value True so it prints NaN
f
```

```python
f = pd.read_csv('F:\\dataTut24(5).csv' , na_filter=False)                      # By default value True so it prints NaN
f
```
