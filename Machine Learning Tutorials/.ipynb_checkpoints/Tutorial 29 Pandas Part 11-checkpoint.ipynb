<center><h1>fillna()</h1></center> <br>
<h3>fillna() - fillna fills the NaN values with given values(input)</h3> <br>
<h2> Syntax - DataFrame.fillna() </h2>

```python
import pandas as pd
```

```python
f = pd.read_csv('F:\\Tutorial25data\\dataTut29.csv')
f
```

# Parameter Value <br>
Give Values to fill NaN values

```python
f.fillna(0)
```

```python
f.fillna(5)
```

# Fill different value in each column use dict

```python
f.fillna({'Name' : 'Beast' , 'Profit' : '40%' , 'Growth': '80%'})
```

# Parameter - Method <br>
ffill - Used to fill previous values <br>
pad  -  Same as ffill

```python
f
```

```python
f.fillna(method = 'ffill')
```

```python
f.fillna(method='pad')
```

# Parameter - Method <br>
bfill - Used to fill backward values <br>

```python
f
```

```python
f.fillna(method='bfill')
```

# Parameter - Axis <br>
0 - For Row(index) <br>
1 - For Column <br>

Note : - **Must Apply value or method fn of fillna , otherwise it gives error**

```python
f.fillna(axis=0)
```

```python
f
```

```python
f.fillna(method = 'ffill' , axis=0)                          # Fill rows by forward values
```

```python
f
```

```python
f.fillna(method = 'ffill' , axis=1)                          # Fill Cols by forward values
```

```python
f.fillna(method = 'bfill' , axis=1)                          # Fill cols by nackward values
```

# Parameter - limit <br>
limit - Used When to fill specific(2,3,4) nan values in single column <br>

```python
f1 = pd.read_csv('F:\\Tut28.csv')
f1
```

# Used limit with value parameter

```python
f1.fillna(value = 2 , limit=1)                                # Only fill one null value for each column because limit=1
```

```python
f1.fillna(value = 2 , limit=2)                                # Only fill one null value for each column because limit=2
```

# Used limit with method parameter

```python
f1
```

```python
f1.fillna(method='ffill', limit=1)                                # Only fill one null value for each column because limit=2
```

```python
f1
```

# Parameter - inplace <br>
inplace - Used replace bool values with NaN  values <br>

Note - inplace didn't create new df, it modifyies original dataframe 

```python
f1
```

```python
f1.fillna(value=5,inplace=True)
f1
```
