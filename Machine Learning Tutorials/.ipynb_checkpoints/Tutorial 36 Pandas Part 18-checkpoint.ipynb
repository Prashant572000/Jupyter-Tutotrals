```python
import pandas as pd
```

```python
s1 = pd.Series([0,1,2])                                              # Create 2 series s1 , s2
s2 = pd.Series([3,4,5])
```

```python
pd.concat([s1,s2])                                                  # Always gives i/p in list , concat 2 series
```

```python
s3 = pd.Series([0,1,2])                                              # Create 2 series s3 , s4
s4 = pd.Series([3,4,5,6,7])

pd.concat([s3,s4])                                                   # Concat unequal series s3 , s4 
```

# Now Create DataFrame & Concat

```python
df1 = pd.DataFrame({'ID' : [1,2,3,4] , 
                   'Name': ['A' , 'B' , 'C' , 'D'],
                   'Class': [5,6,7,8]})
df2 = pd.DataFrame({'ID' : [5,6,7,8] , 
                   'Name': ['E' , 'F' , 'G' , 'H'],
                   'Class': [9,10,11,12]})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])                                                # Concat 2 data frames row wise(axis=0) By Default
```

```python
pd.concat([df2,df1])                                              # Change position now upper df2 , below df1
```

```python
pd.concat([df1,df2] , axis=1)                                     # Concat 2 df by column(axis=1) By default axis=0(Row) 
```

```python
pd.concat([df1,df2] , axis=0)                                   # Here axis=0 optional because it is by default
```

```python
pd.concat([df1,df2] , ignore_index=True)          #But index start after new df to fix this use fn ignore_index=False By default
```
