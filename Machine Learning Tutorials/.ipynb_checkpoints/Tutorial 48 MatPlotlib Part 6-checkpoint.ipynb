```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of IOT Subjects, classes
# Classes = ['Python' , 'R' , 'Artificial Intelligence' , 'Machine Learning' , 'Data Science']

classes = ['Python' , 'R' , 'AI' , 'ML' , 'DS']
class1_students = [30,10,20,25,10]              # Out of 100 student in each class
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
plt.bar(classes , class1_students , width=0.7, align="edge", color='k',edgecolor='m',linewidth=5,alpha=0.6,linestyle='--'
       , label='class1_students') # visible=False
plt.legend()
```

```python
# Change background color of chart
style.use("ggplot")

plt.bar(classes , class1_students , width=0.7, align="edge", color='k',edgecolor='m',linewidth=5,alpha=0.6,linestyle='--'
       , label='class1_students') # visible=False
plt.legend()
```

```python
# Change size of chart using figure
plt.figure(figsize=(16,9))
plt.bar(classes , class1_students , width=0.7, align="edge", color='k',edgecolor='m',linewidth=5,alpha=0.6,linestyle='--'
       , label='class1_students') # visible=False
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.barh(classes , class1_students , align="edge", color='k',edgecolor='m',linewidth=5,alpha=0.6,linestyle='--'
       , label='class1_students') # visible=False
style.use("ggplot")
plt.legend()# Show chart Horizontally 
```

```python
# Give x,y label , title to chart 
plt.figure(figsize=(16,9))
plt.bar(classes , class1_students , width=0.7, align="edge", color='k',edgecolor='m',linewidth=5,alpha=0.6,linestyle='--'
       , label='class1_students') # visible=False
plt.title("Bar Chart of IAIP Class" , fontsize=15)
plt.xlabel("Classes" , fontsize=12)
plt.ylabel("Number of Students" , fontsize=12)
plt.legend()
```

```python
# chart of Mutiple data 

plt.bar(classes , class1_students,  color='g' , label='class1_students') 
plt.bar(classes , class2_students ,  color='b', label='class2_students') 
plt.title("Bar Chart of IAIP Class" , fontsize=15)
plt.xlabel("Classes" , fontsize=12)
plt.ylabel("Number of Students" , fontsize=12)
plt.legend()
```

```python
# But above chart overlap to each other, make seperate chart
classes_index=np.arange(len(classes))
width=0.4
plt.bar(classes_index , class1_students,width,  color='g' , label='Class 1 Students') 
plt.bar(classes_index+width , class2_students,width,  color='b' , label='Class 2 Students')
plt.title("Bar Chart of IAIP Class" , fontsize=15)
plt.xlabel("Classes" , fontsize=12)
plt.ylabel("Number of Students" , fontsize=12)
plt.legend()
```

```python
# To show bar values use xticks fn
classes_index=np.arange(len(classes))
width=0.4
plt.bar(classes_index , class1_students,width,  color='g' , label='Class 1 Students') 
plt.bar(classes_index+width , class2_students,width,  color='b' , label='Class 2 Students')
plt.title("Bar Chart of IAIP Class" , fontsize=15)
plt.xlabel("Classes" , fontsize=12)
plt.ylabel("Number of Students" , fontsize=12)
plt.xticks(classes_index+width , classes)                                   # classes(Py,R,AI,ML) replace class_index(0,1,2,3) 
plt.legend()
```

```python
# To rotate class values
classes_index=np.arange(len(classes))
width=0.4
plt.bar(classes_index , class1_students,width,  color='g' , label='Class 1 Students') 
plt.bar(classes_index+width , class2_students,width,  color='b' , label='Class 2 Students')
plt.title("Bar Chart of IAIP Class" , fontsize=15)
plt.xlabel("Classes" , fontsize=12)
plt.ylabel("Number of Students" , fontsize=12)
plt.xticks(classes_index+width , classes , rotation=30)                                    
plt.legend()
```

```python
# To Show all value
classes_index=np.arange(len(classes))
width=0.2                                        # Only took 1%(100) , we took 0.2+0.2+0.2 , only 40 % left
plt.bar(classes_index , class1_students,width,  color='g' , label='Class 1 Students') 
plt.bar(classes_index+width , class2_students,width,  color='b' , label='Class 2 Students')
plt.bar(classes_index+width+width , class3_students,width,  color='y' , label='Class 3 Students')

plt.title("Bar Chart of IAIP Class" , fontsize=15)
plt.xlabel("Classes" , fontsize=12)
plt.ylabel("Number of Students" , fontsize=12)
plt.xticks(classes_index+width , classes , rotation=30)                                    
plt.legend()
```

```python
# Show above graph Horizontally , oppsite x,y label , xticks, plt.bar
classes_index=np.arange(len(classes))
plt.figure(figsize=(16,9))
width=0.2
plt.barh(classes_index , class1_students , width , color='g' , label='Class 1 Students')
plt.barh(classes_index+width , class2_students , width , color='b' , label='Class 2 Students')
plt.barh(classes_index+width+width , class3_students , width , color='y' , label='Class 3 Students')
plt.title("Bar Chart of IAIP Class" , fontsize=25)
plt.yticks(classes_index+width , classes , rotation=30)                                    
plt.ylabel("Classes" , fontsize=20)
plt.xlabel("No of Students" , fontsize=20)
plt.legend()
```
