```python
x=f["Rating"]
y=f["Reviews"]                                                      # Assign rating & reviews to x & y to plot graph
```

```python
plt.scatter(x,y)                                                   # Draw scatter plot
```

```python
plt.scatter(x,y)                                                   # Give title , labels to chart
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.scatter(x,y , c="r")                                             # Give color to red
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))                                        # Increse figure size
plt.scatter(x,y , c="r")                                             
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))                                        
plt.scatter(x,y , c="r" , marker="*")                              # Give marker, dot replace with *
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))                                        
plt.scatter(x,y , c="r" , marker="*" , s=100)                                # increase marker size 
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))                                        
plt.scatter(x,y , c="r" , marker="*" , s=100 , alpha=0.3)                                # Change Transparency/Opacity 
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))                                        
plt.scatter(x,y , c="r" , marker="*" , s=100, linewidth=7)                                # increase * size 
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))                                        
plt.scatter(x,y , c="r" , marker="*" , s=100 , linewidth=7 , edgecolor='y')                        # Change * line color 
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))                                        
plt.scatter(x,y , c="r" , verts='<' , s=100 , linewidth=7 , edgecolor='g')             # use verts as marker, both same 
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Reviews" , fontsize=15)
```

# Plot multiple data in single chart

```python
plt.scatter(x,y , c="r" , marker='*' , s=50 , linewidth=5 , edgecolor='g')
plt.scatter(x,f["Installs"] , c="b" , marker='o' , s=50 , linewidth=5 , edgecolor='c')
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Installs" , fontsize=15)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y , c="r" , marker='*' , s=50 , linewidth=5 , edgecolor='g')
plt.scatter(x,f["Installs"] , c="b" , marker='o' , s=50 , linewidth=5 , edgecolor='c')
plt.title("Google Play Store Apps Data", fontsize=20)
plt.xlabel("Rating" , fontsize=15)
plt.ylabel("Installs" , fontsize=15)
```
