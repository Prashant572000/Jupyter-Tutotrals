```python
textdop={"fontsize":15}
colores=['c' , 'y' , 'b' , 'm' , 'g']
explode=[0.1,0,0.2,0,0]
plt.pie(class1_students , labels=classes , explode=explode , colors=colores , autopct="% 0.1f%%"  , shadow=True, radius=1.5,
        startangle=270,textprops=textdop)
plt.legend()                                                            # Give legend to Chart
plt.show()
```

```python
textdop={"fontsize":15}
colores=['c' , 'y' , 'b' , 'm' , 'g']
explode=[0.1,0,0.2,0,0]
plt.pie(class1_students , labels=classes , explode=explode , colors=colores , autopct="% 0.1f%%"  , shadow=True, radius=1.5,
        startangle=270,textprops=textdop)
plt.legend(loc=5)                                                    # Set location of legend
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(6,2))

colors=['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels=np.ones(20)                                                        # Create 1D array 20 elements - 1

plt.pie([1] , colors='k' , radius=2.05)                                   # Plot outer black circle
plt.pie(labels, colors=colors, radius=2.0)                                # Plot red,white circle using colors(20) 

plt.pie([1] , colors='g' , radius=1.8)                                    # Plot green line  with radius
plt.pie([1] , colors='y' , radius=1.6)                                    # Plot yellow line with radius                      
plt.pie([1] , colors='c' , radius=1.3)                                    # Plot Shine color line 
plt.pie([1] , colors='b' , radius=1.1)                                    # Plot blue color line
plt.pie([1] , colors='m' , radius=0.9)                                    # Plot Magenda color line

plt.pie([1] , colors='b' , radius=0.31)                                   # Plot blue line
plt.pie(labels , colors=colors , radius=0.3)                              # Plot red , white line

plt.pie([1] , colors='w' , radius=0.2)                                    # Plot white line
plt.pie([1] , colors='k' , radius=0.1)                                    # Plot black circle

plt.show()
```

```python
xdata=[22,33,44,55,66]                              # Give data
dn = ['T1','T2','T3','T4','T5']                     # Give labels

plt.pie(xdata,labels=dn , radius=1.3, textprops={"fontsize":15}, autopct="% 0.3f%%" , counterclock=False,rotatelabels=True ,
       center=(4,0))
plt.legend(loc=5)
plt.show()
```
