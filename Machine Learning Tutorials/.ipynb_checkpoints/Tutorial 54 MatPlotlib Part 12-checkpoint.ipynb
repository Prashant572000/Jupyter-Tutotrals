<center><h1>Imshow & Colorbar Part 1</h1></center> <br>
How to show image using matplotlib imgshow fn

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg             # Used to read , show image & perform some specific operations 
```

```python
img1 = mpimg.imread('Tut53(1).png')          # Read image using imread fn(Takes 2 argu, fname,format)& store in img1 variable 
```

```python
img1                                         # Readed image is stored in numpy 3D array  
```

```python
type(img1)                                                   # img data stored using numpy array
```

```python
img1.shape                                                   # Check Shape of imh=g1
```

```python
img1.ndim                                                    # It is 3 Dim because img is in RGB (Color) format
```

```python
plt.imshow(img1)                                             # Show img data
```

```python
plt.imshow(img1) 
plt.show()                                                     # Don't show above information
```

```python
plt.axis("off")                                                 # To axis off (By Default ON)
plt.imshow(img1) 
plt.show()                                                     
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")                                                 
plt.imshow(img1) 
plt.show()                                                     
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img1)
plt.colorbar()                                                              # Show colorbar of image (Don't need parameter)
plt.show()
```

```python
plt.axis("off")                                                 # To axis off (By Default ON)
plt.imshow(img1 , cmap="Accent")                                # cmap only works when img data is 2D, but we have 3D
plt.colorbar()
plt.show()                                                     
```

```python
single_channel = img1[:,:,1]                                  # Take 3 out of 2nd channel's data 
plt.axis("off")                                               # To axis off (By Default ON)
plt.imshow(single_channel , cmap="hot")                    # Give single_channel, then cmap affects on img1
plt.colorbar()
plt.show()                                                     
```

```python
single_channel = img1[:,:,2]                                  # Take 3 out of 3rd channel's data 
plt.axis("off")                                               # To axis off (By Default ON)
plt.imshow(single_channel , cmap="hot")                    # Give single_channel, then cmap affects on img1
plt.colorbar()
plt.show()                                                     
```

```python
single_channel = img1[:,:,0]                                  # Take 3 out of 1st channel's data 
plt.axis("off")                                               # To axis off (By Default ON)
plt.imshow(single_channel , cmap="hot")                    # Give single_channel, then cmap affects on img1
plt.colorbar()
plt.show()                                                     
```
