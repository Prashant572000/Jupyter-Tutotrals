```python
df1 = sns.load_dataset("tips")                                      # Load default seaborn dataset Online
df1
```

```python
sns.lineplot(x="size" , y="total_bill" , data=df1 )
```

```python
sns.lineplot(x="size",y="total_bill",data=df1, hue="sex")                                        # Use hue
```

```python
sns.lineplot(x="size",y="total_bill",data=df1, hue="sex", style="sex")                               # Use style
```

```python
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex", style="sex" , palette="Accent")  # palette same as colormap of matlib
```

```python
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex", style="sex" , palette="hot")  # palette same as colormap of matlib
```

```python
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex", style="sex" , palette="Accent",
            dashes=False)                 # Give (BY DEFAULT)True for dashed line , else False   
```

```python
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex", style="sex" , palette="hot"
            , dashes=False , markers=["o" , "<"])                     # Give marker Male (Circle), Female (Triangle)  
```

```python
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex", style="sex" , palette="hot",
            markers=['o','<'] , legend='full')
                                        # Give legend to False(not show legend) , (brief(Default), or Full(To show legend))
```

```python
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex",style="sex",palette="hot", markers=['o','<'],
            dashes=False)
plt.title("Seaborn Line Plot", fontsize=20)
plt.xlabel("Size" , fontsize=16)
plt.ylabel("Total Bill" , fontsize=16)
```

```python
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex",style="sex",palette="hot", markers=['o','<'],
            dashes=False)
plt.title("Seaborn Line Plot", fontsize=20)
plt.xlabel("Size" , fontsize=16)
plt.ylabel("Total Bill" , fontsize=16)
plt.show()                                                        # To hide above information
```

```python
sns.set(style='darkgrid')                                                                      # Set background color 
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex",style="sex",palette="hot", markers=['o','<'],
            dashes=False)
plt.title("Seaborn Line Plot", fontsize=20)
plt.xlabel("Size" , fontsize=16)
plt.ylabel("Total Bill" , fontsize=16)
plt.show()                                                        
```

```python
plt.figure(figsize=(16,9))
sns.set(style='darkgrid')                                                                       
sns.lineplot(x="size",y="total_bill",data=df1,hue="sex",style="sex",palette="hot", markers=['o','<'],
            dashes=False)
plt.title("Seaborn Line Plot", fontsize=20)
plt.xlabel("Size" , fontsize=16)
plt.ylabel("Total Bill" , fontsize=16)
plt.show()   
```

```python
sns.set(style='darkgrid')                                                                      # Set background color 
sns.lineplot(x="size",y="total_bill",data=df1,hue="day",style="day",palette="hot", markers=['o','<'],
            dashes=False)
plt.title("Seaborn Line Plot", fontsize=20)
plt.xlabel("Size" , fontsize=16)
plt.ylabel("Total Bill" , fontsize=16)
plt.show()        # Below gives error bcz we give hue,style to day, and marker size(2)!=day(4 , Sun,Thur,Fir,Sat)
```

```python
sns.set(style='darkgrid')                                                                      # Set background color 
sns.lineplot(x="size",y="total_bill",data=df1,hue="day",style="day",palette="hot", markers=['o','<','>','^'],
            dashes=False)
plt.title("Seaborn Line Plot", fontsize=20)
plt.xlabel("Size" , fontsize=16)
plt.ylabel("Total Bill" , fontsize=16)
plt.show()   
```
