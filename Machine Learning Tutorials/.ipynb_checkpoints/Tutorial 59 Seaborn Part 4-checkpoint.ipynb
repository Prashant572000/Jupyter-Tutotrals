Displot fn plots histogram by default

```python
import seaborn as sns
from scipy.stats import norm                                      # Used to plot normalize line
import matplotlib.pyplot as plt
```

```python
df1 = sns.load_dataset("tips")                                    # Load seaborn's default data set (Online)
df1
```

```python
sns.distplot(df1['size'])                                                    # Plot numeric data using distplot
```

Above we get histogram with lineplot is called KDE(Kernel Density Estimator) 

```python
sns.distplot(df1['tip'])                                         # Show histogram of tip col
```

```python
sns.distplot(df1['total_bill'])
```

```python
"""
sns.distplot(
    a,               Give data(col) to show plot
    bins=None,       Show data only in specified no
    hist=True,       To hide hist
    kde=True,        To hide kde
    rug=False,       Shows vertical bar on data
    fit=None,        fit normalize line, need to import norm which is in scipy library
    hist_kws=None,
    kde_kws=None,
    rug_kws=None,
    fit_kws=None,
    color=None,      Change histogram color 
    vertical=False,  show horizontal plot
    norm_hist=False, less usable
    axlabel=None,    Change xlabel heading
    label=None,      Give label working only in 
    ax=None,
)
""";
```

```python
sns.distplot(df1['total_bill'] , bins=[0,10,15,30,50,60])                # Plot histogram of total_bill with bins 
                                                          # To show bins on x axis , use xticks parameter also
```

```python
sns.distplot(df1['total_bill'] , hist=False)                   # By default hist=True(To hide histogram)
```

```python
sns.distplot(df1['total_bill'] , kde=False)             # To hide kde, show histogram GIVE kde=False(Default True) 
```

```python
sns.distplot(df1['total_bill'] , rug=True)        # data shows in which location it shows vertical bar , use rug(Default False)
```

```python
sns.distplot(df1['total_bill'], fit=norm)                  # It gives normalize line
```

```python
sns.distplot(df1['total_bill'], fit=norm , kde=False)                  # It gives normalize line, hide kde
```

```python
sns.distplot(df1['total_bill'], color='r')                  # It gives red color to histogram
```

```python
sns.distplot(df1['total_bill'], color='r' , vertical=True)              #show histogram plot horizontally Default(ver=False)
```

```python
sns.distplot(df1['total_bill'], norm_hist=True)                  # norm_hist(default False), less usable
```

```python
sns.distplot(df1['total_bill'], axlabel='Total Bill')                  # Change xlabel heading
```

```python
sns.distplot(df1['total_bill'], label="Total Bill")                  # To give label, working when give legend using plt
plt.legend()
```

```python
sns.distplot(df1['total_bill'], label='Total Bill')
plt.title("Histogram Bill Plot")                                  # Give plot
plt.legend()
```
