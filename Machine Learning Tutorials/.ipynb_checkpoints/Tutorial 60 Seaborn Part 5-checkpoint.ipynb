```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df = sns.load_dataset("tips")                                       # load and print dataset
df
```

```python
sns.set()                                                 # By defaut style = darkgrid (Change bg color)
sns.distplot(df['total_bill'], label='Total Bill')
plt.title("Histogram Plot of Total Bill")
plt.legend()
```

```python
plt.figure(figsize=(16,9))                                        # Increase size of graph
sns.set()   
sns.distplot(df['total_bill'], label='Total Bill')
plt.title("Histogram Plot of Total Bill")
plt.legend()
```

# Sort Particular columns data

```python
df.total_bill.sort_values()                            # Sort any columns data in file
```

# Use bins parameter

```python
sns.set()
sns.distplot(df['total_bill'], bins=[1,5,10,15,20,25,30,35,40,45,50,55], label='Total Bill', axlabel='Total Bill')
plt.title("Histogram line plot" , fontsize=20)
plt.legend(fontsize=15)
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]                       # bins only used with xticks 
sns.distplot(df['total_bill'] , bins=bins)
plt.xticks(bins)
plt.title("Histogram of Total Bill")
```

```python
plt.xticks(bins)
sns.distplot(df['total_bill'] , hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00'
            , 'linewidth':5, 'linestyle':'--' , 'alpha':0.6})             # use hist keywords
sns.set()
plt.title("Histogram of Total Bill")
```

```python
sns.distplot(df['total_bill'],hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--','alpha':0.8},
kde_kws = {'color':'#8e00ce', 'linewidth':8, 'linestyle':'--' , 'alpha':0.5},
            rug=True,
            rug_kws={'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':2, 'linestyle':'--','alpha':0.9})   
```

# While use fit parameter , set kde = False , else can't used

```python
sns.distplot(df['total_bill'],hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--','alpha':0.8},
kde_kws = {'color':'#8e00ce', 'linewidth':8, 'linestyle':'--' , 'alpha':0.5},
            rug=True,
            rug_kws={'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':2, 'linestyle':'--','alpha':0.9},
            fit=norm,
          fit_kws={'color':'#8e00ce', 'linewidth':12, 'linestyle':'--' , 'alpha':0.4})   
```

```python
plt.figure(figsize=(16,9))
sns.distplot(df['total_bill'],hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00', 'linewidth':5, 'linestyle':'--','alpha':0.8},
kde_kws = {'color':'#8e00ce', 'linewidth':8, 'linestyle':'--' , 'alpha':0.5},
            rug=True,
            rug_kws={'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':2, 'linestyle':'--','alpha':0.9},
            fit=norm,
          fit_kws={'color':'#8e00ce', 'linewidth':12, 'linestyle':'--' , 'alpha':0.4},label='TB')
plt.title("Histogram Plot Total Bill" , fontsize=25)
plt.xlabel("Total Bill" , fontsize=20)
plt.legend(fontsize=15)
```
