```python
df = sns.load_dataset("titanic")                               # Load and print default dataset
df
```

```python
sns.scatterplot(x='age' , y='fare' , data=df )                                  # Give x & y value to scatterplot
```

```python
sns.scatterplot(x='age' , y='fare', data=df, hue='sex')                         # Give hue parameter with x&y
```

```python
sns.scatterplot(x='age' , y='fare' , data=df, hue='sex', style='who')           # Give who col to style
```

```python
sns.scatterplot(x='age',y='fare',data=df,hue='sex',style='who',size='who')          # To adjust size of scatter plot use SIZE
```

# Increase size of dots use sizes(take i/p as tuple) sizes=(min val , max val)

```python
sns.scatterplot(x='age',y='fare',data=df,hue='sex',style='who',size='who', sizes=(100,300))
```

```python
plt.figure(figsize=(16,9))                                                                  # Increse size of plot
sns.scatterplot(x='age',y='fare',data=df,hue='sex',style='who',size='who',sizes=(100,300))
```

In titanic which more survive , Man , Woman , Child

```python
sns.scatterplot(x='who',y='fare',data=df,hue='alive',style='alive',size='who',sizes=(100,200))
```

```python
sns.scatterplot(x='who',y='fare',data=df,hue='alive',style='alive',size='who',sizes=(100,200)
               , palette='inferno_r' , alpha=0.8)                              # Use palette & alpha parameter
```

```python
plt.figure(figsize=(10,8))                                # Create effective Plot with bar, scatter, line plot
sns.scatterplot(x='age', y='fare' , data=df)
sns.lineplot(x='age', y='fare', data=df)
sns.barplot(x='age' , y='fare' , data=df)
plt.title("Mupltiple Plots " , fontsize=30)
plt.xlabel("Age", fontsize=25)
plt.ylabel("Fare" , fontsize=25)
plt.show()
```
