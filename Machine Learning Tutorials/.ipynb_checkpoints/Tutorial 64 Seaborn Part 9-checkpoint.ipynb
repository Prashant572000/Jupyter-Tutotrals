<center> <h1>Seaborn Heatmap Part 1</h1></center> <br>
It is a graphical representation of 2D data. Each data value represents in matrix & has a special color.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
arr2d = np.linspace(1,5,12).reshape(4,3)               # Generate 12 no. b/w 1-5, reshape 4R X 3C 2D data
arr2d
```

```python
sns.heatmap(arr2d)                                     # It will generate heatmap 
```

```python
data1 = pd.read_csv('E:\Coding Tutorials\Jupyter Tutorial\Machine Learning Tutorials\Tutorialcsvdata\Tut64.csv')
data1
```

# Drop Some Columns and make 1 column index

```python
data2 = data1.drop(columns=['Country Code' , 'Indicator Name', 'Indicator Code'] , axis=1).set_index('Country Name')
data2
```

```python
sns.heatmap(data2)                                             # Generate heatmap of new data2
```

```python
plt.figure(figsize=(16,9))                                       # Set colorbar's min & max value 
sns.heatmap(data2 , vmin=0 , vmax=21)
```

```python
plt.figure(figsize=(16,9))                                         # Set colormap By giving values
sns.heatmap(data2 , vmin=0 , vmax = 20 , cmap='coolwarm')          # Cool means - Blue , Warm means - Red
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(data2, vmin=0,vmax=22,cmap='coolwarm',annot=True)                         # Give actual value upon color
```
