```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data1 = pd.read_csv('E:\Coding Tutorials\Jupyter Tutorial\Machine Learning Tutorials\Tutorialcsvdata\Tut64.csv') # read csv
data2 = data1.drop(columns=['Country Code' , 'Indicator Name', 'Indicator Code'] , axis=1).set_index('Country Name')
data2                                                          # drop some str column, and set index - Country name
```

```python
plt.figure(figsize=(10,8))
annot_keywords = {'fontsize':10, 'fontstyle':'italic','color':'k', 'alpha':0.9,'rotation':'vertical',
            'verticalalignment':'center', 'backgroundcolor':'w'}           # Give some basic annot keywords to annot_kws fn 
sns.heatmap(data2 ,cmap='coolwarm',annot=True, annot_kws = annot_keywords)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(data2, cmap='coolwarm', annot=True, linewidth=5)                     # Show seperate box of each element
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data2 , cmap='coolwarm' ,annot=True, linewidth=5 , linecolor='k')                          # Set line color
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data2, annot=True, linewidth=5 , linecolor='k' , cmap='coolwarm' , cbar=False)           # Hide colorbar
```

```python
plt.figure(figsize=(16,9))                                                              # Hide X axis & Y axis labels
sns.heatmap(data2, annot = True , cbar=False, xticklabels=False , yticklabels=False) 
```

# Show Colorbar on different - 2 Format 

```python
plt.figure(figsize=(16,9))                  
colorbar_arguments = {'orientation':'horizontal',
                     'shrink' : 1 ,                # To increase colorbar size
                      'extend' : 'min', 'extendfrac':0.1, 'ticks' : np.arange(0,22),  # Colrbar range from 0 - 21
                      'drawedges' : True}   # To show edges(Line by Line)
sns.heatmap(data2, annot = True ,  cbar_kws=colorbar_arguments) 
```

```python
plt.figure(figsize=(16,8))                                                        # Set x and y label headings manually
country_code = ['USA','GBR','IND','CHN','RUS','AUS','FRA','DEU','CAN','BRA','ARG','PAK','NEP','BAN','JAP']
sns.heatmap(data2 , annot=True , xticklabels = np.arange(0,15) , yticklabels=country_code)
```

```python
plt.figure(figsize=(16,8))                                                        # Set Title , labels
ax = sns.heatmap(data2 , annot=True)
ax.set(title='Heatmap of Who Is Responsible for Global Warming' , 
      xlabel='CO2 Emissions Per Year',
      ylabel='Country Name',)
```

```python
plt.figure(figsize=(20,9))                                                        
ax = sns.heatmap(data2)
ax.set(title='Heatmap of Who Is Responsible for Global Warming' , 
      xlabel='CO2 Emissions Per Year',
      ylabel='Country Name',)
sns.set(font_scale=5)                                                # Set font size
```
