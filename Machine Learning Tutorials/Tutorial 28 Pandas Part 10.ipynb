<center> <h1> Handling Missing Values in Python </h1> </center>

dropna() - It drops nan, null, missing values Row OR Col , And creates a new DataFrame , where no null vals 

```python
import pandas as pd
```

```python
f1 = pd.read_csv("F:\\Tut28.csv")
f1
```

```python
f1.dropna()                                                           # It drop whole rows where nan val found
```

# Axis Parameter

```python
f1.dropna(axis=1)                                             # Remove full column, where single nan value found
```

```python
f1.dropna(axis=0)                                             # Remove full row, where nan value found
```

# How Parameter - 2 types of I/p <br>
Any - Drop row/col where it find any nan values. By Default<br>
All - Drop row/col where it find all nan values in row and col.<br>

```python
f1.dropna(how='any')
```

```python
f1.dropna(how='all')
```

```python
f1 = pd.read_csv('F:\\Tut28.csv')
f1
```

```python
f1.dropna(how='all')
```

# thresh - We give integer as i/p By default val None <br>
It is used when we want minimum nan values in row & col

```python
f1.dropna(thresh=1)
```

# If we didn't give Axis Parameter By default it takes Axis=0(Row)

```python
f1.dropna(thresh=4)                                           # It prints row/col where it finds min 4 notnull values
```

```python
f1.dropna(thresh=5)                                           # It prints row/col where it finds min 5 notnull values
```

```python
f1 = pd.read_csv("F:\\Tut28.csv")
f1
```

# Subset - It is used when drop specific row/col , or nan values row & col. It is Optional 

```python
f1.dropna(subset = ['Name'])
```

```python
f1.dropna(subset = ['Profit'])
```

# Inplace - If give True, it update that df & drop nan values, not create newdf <br>
we give i/p in boolean format 

```python
f1 = pd.read_csv("F:\\Tut28.csv")
f1
```

```python
f1.dropna(inplace=True)
f1
```

```python
f1.dropna(inplace=False)
f1
```
