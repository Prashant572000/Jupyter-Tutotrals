<center><h1>interpolate()</h1></center> <br>
By using interpolate we can fill NaN values with DataFrame OR Series <br>
Powerful Fn  <br>
Use Various interpolation techniques() Ex - linear method <br>

Syntax - DataFrame.interpolate() <br>
         Series.interpolate() <br>

```python
import pandas as pd
```

```python
f = pd.read_csv('E:\\Coding Tutorials\\Jupyter Tutorial\\Machine Learning Tutorials\\Tutorialcsvdata\\Tut31.csv')
f.head(10)
```

```python
f.interpolate()                                       # It guess and fill only numeric values
```

```python
f.interpolate(method='linear')
```

```python
f.interpolate(method='time')                                 # It gives error because date is string type 
```

```python
type(f.Date[0])
```
