<center><h1>Concat() fn   Part 2</h1></center><br>
We can combine together Series,dataframe objects by merge , concat , join fns<br>
Syntax - df.concat() <br>
Parameters : - <br>
    objs - <br>
    axis - <br>
    join - <br>
    ignore_index - <br>
    keys - <br>
    sort - <br>

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID' : [1,2,3,4] , 
                   'Name': ['A' , 'B' , 'C' , 'D'],
                   'Class': [5,6,7,8]})
df2 = pd.DataFrame({'ID' : [3,4] , 
                   'Name': ['C' , 'D'],
                   'Class': [7,8]})
df1
```

```python
df2
```

```python
pd.concat([df1,df2])                                                    # Concat dfs row(axis=0) wise by default
```

```python
pd.concat([df1,df2] , axis=1)                                                    # Concat dfs col(axis=1) wise
```

# Parameter <br>

Join - Takes 2 i/p - <br>
inner - It returns intersection(Both values common in df1,df2) <br>
outer - returns union of df1 ,df2 (All values of df1,df2)       BY DEFAULT


```python
pd.concat([df1,df2] , join='inner')
```

```python
pd.concat([df1,df2] , join='inner' , axis=1)
```

# Parameter <br>

Keys - BY DEFAULT None <br>

It seperates multiple dataframes
       


```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2] , keys=['df1' , 'df2'])                  # keys=[](in list give labels(col) in which we seperate 2 dfs)
```

```python
pd.concat([df1,df2] , keys=['First df' , 'Secondf2'] , axis=1)  # keys=[](in list give labels(col) in which we seperate 2 dfs)
```

# If we have multiple dataframes with diffrent-2 columns Then

```python
df3 = pd.DataFrame({'ID' : [1,2,3,4] , 
                   'Name': ['A' , 'B' , 'C' , 'D'],
                   'Class': [5,6,7,8]})
df4 = pd.DataFrame({'Marks' :[30,63,91,35]})
df3
```

```python
df4
```

```python
pd.concat([df3,df4])
```

```python
pd.concat([df3,df4], ignore_index=True)
```
