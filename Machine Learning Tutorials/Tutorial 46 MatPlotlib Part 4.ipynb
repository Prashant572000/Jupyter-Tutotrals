```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45,(100))            # Generate ml stu information b/w 18 & 45,2nd argu- Total100 student
py_student_age = np.random.randint(15,40,(100))            # Generate py stu information b/w 15 & 40,2nd argu- Total100 student
display(ml_student_age,py_student_age)
```

# Show this information on bar is called Histogram

```python
plt.hist(ml_student_age)                            # Give data to show histogram
plt.title("Machine Learning Student Histograms")
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
plt.hist(ml_student_age , bins=[15,20,25,30,35,40,45])              # Define category of graph use bins
plt.title("Machine Learning Student Histograms")
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.show()
```

# To show bar properly use rwidth <br> 
By default 1(100%) <br>
Means it takes 100% width to show bar <br>
Set 0.8(80%) to show data, 0.2(20%) empty space


```python
plt.hist(ml_student_age , bins=[15,20,25,30,35,40,45], rwidth=0.8)            
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.show()
```

# Use histtype parameter Takes 4 argu - bar(Default), barstacked,step,stepfilled <br>

```python
plt.hist(ml_student_age , bins=[15,20,25,30,35,40,45], rwidth=0.8 , histtype="barstacked")            
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.show()
```

# Use align parameter Takes 3 argu - mid(Default), left,right<br>

```python
plt.hist(ml_student_age , bins=[15,20,25,30,35,40,45], rwidth=0.8 , histtype="bar" , align='right')            
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# Use orientation parameter Takes 2 argu - Vertical(Default) ,Horizontal<br>
plt.hist(ml_student_age , bins=[15,20,25,30,35,40,45], rwidth=0.8 , histtype="bar", orientation='horizontal')            
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# To Change color of bar use - color 
plt.hist(ml_student_age,bins=[15,20,25,30,35,40,45], rwidth=0.8,histtype="bar",align='right',color='m',orientation='horizontal')
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.show()
```

```python
# To show labels, use label with legend also(Else not work)

plt.hist(ml_student_age,bins=[15,20,25,30,35,40,45], rwidth=0.8,histtype="bar",align='right',color='m',
        orientation='horizontal' , label="ML Student")
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

# Show Multiple Graph Student

```python
plt.hist(ml_student_age,bins=[15,20,25,30,35,40,45], rwidth=0.8,histtype="bar",align='right',color='m',
        orientation='horizontal' , label="ML Student")
plt.hist(py_student_age,bins=[10,15,20,25,30,35,40], rwidth=0.8,histtype="bar",align='right',color='y',
        orientation='horizontal' , label="Py Student")
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

```python
# Show ml student Horizontal, py student Vertical 
plt.hist(ml_student_age,bins=[15,20,25,30,35,40,45], rwidth=0.8,histtype="bar",align='right',color='m',
        orientation='horizontal' , label="ML Student")
plt.hist(py_student_age,bins=[10,15,20,25,30,35,40], rwidth=0.8,histtype="bar",align='right',color='y',
         label="Py Student")
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

```python
# Show ml student Horizontal, py student Horizontal 
plt.hist(ml_student_age,bins=[15,20,25,30,35,40,45], rwidth=0.8,histtype="bar",align='right',color='m',
        orientation='vertical' , label="ML Student")
plt.hist(py_student_age,bins=[10,15,20,25,30,35,40], rwidth=0.8,histtype="bar",align='right',color='y',
         orientation='vertical' ,label="Py Student")
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

```python
# Plot Bar seperately
plt.hist([ml_student_age, py_student_age],bins=[15,20,25,30,35,40,45], rwidth=0.8,histtype="bar",color=['m','y'],
         label=["ML Student", "Py Student"])
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```

```python
# To increase histogram size, use plt.figure(figsize=(16,9)) 
plt.figure(figsize=(16,9))                                        # Give Size in Tuple
plt.hist([ml_student_age, py_student_age],bins=[15,20,25,30,35,40,45], rwidth=0.8,histtype="bar",color=['m','y'],
         label=["ML Student", "Py Student"])
plt.title("Machine Learning Student Histograms")
plt.xlabel("Students Age category")
plt.ylabel("No. Students Age")
plt.legend()
plt.show()
```
