Show % value in data use autopct <br>
autopct="%0.2f%% - Show 2 values after . <br>
autopct="%0.1f%% - Show 1 values after.

```python
classes=["Python" , 'R' , 'Artificial Intelligence' , 'Machine Learning' , 'Data Science']
class1_students=[45,15,35,25,30]
explode=[0.1,0,0.2,0,0]
colores=['c','b','r','y','g']                                     
plt.pie(class1_students , labels=classes,explode=explode,colors=colores ,autopct="%0.1f%%")           # Shows % data 
plt.show()
```

```python
classes=["Python" , 'R' , 'Artificial Intelligence' , 'Machine Learning' , 'Data Science']
class1_students=[45,15,35,25,30]
explode=[0.1,0,0.2,0,0]
colores=['c','b','r','y','g']                                     
plt.pie(class1_students , labels=classes,explode=explode,colors=colores ,autopct="%0.2f%%")           # Shows % data 
plt.show()
```

```python
classes=["Python" , 'R' , 'Artificial Intelligence' , 'Machine Learning' , 'Data Science']
class1_students=[45,15,35,25,30]
explode=[0.1,0,0.2,0,0]
colores=['c','b','r','y','g']                                     
plt.pie(class1_students,labels=classes,explode=explode,colors=colores,
        autopct="%0.1f%%" , shadow=True)                                  # Shows shadow Give shadow=True , BY Default False
plt.show()
```

```python
classes=["Python" , 'R' , 'Artificial Intelligence' , 'Machine Learning' , 'Data Science']
class1_students=[45,15,35,25,30]
explode=[0.1,0,0.2,0,0]
colores=['c','b','r','y','g']                                     
plt.pie(class1_students,labels=classes,explode=explode,colors=colores,
        autopct="%0.1f%%" , shadow=True , radius=1.4)                             # To increase radius of chart use radius
plt.show()
```

```python
classes=["Python" , 'R' , 'Artificial Intelligence' , 'Machine Learning' , 'Data Science']
class1_students=[45,15,35,25,30]
explode=[0.1,0,0.2,0,0]
colores=['c','b','r','y','g']                                     
plt.pie(class1_students,labels=classes,explode=explode,colors=colores,
        autopct="%0.1f%%" , shadow=True ,startangle=90)             # To rotete chart , use startangle Give i/p : 0 - 360
plt.show()
```

```python
classes=["Python" , 'R' , 'Artificial Intelligence' , 'Machine Learning' , 'Data Science']
class1_students=[45,15,35,25,30]
explode=[0.1,0,0.2,0,0]
colores=['c','b','r','y','g']                                     
plt.pie(class1_students,labels=classes,explode=explode,colors=colores,
        autopct="%0.1f%%" , shadow=True ,startangle=270)             # To rotete chart , use startangle 0 - 360
plt.show()
```

```python
explode=[0.1,0,0.2,0,0]
colores=['c','b','r','y','g']
textpreporty = {"fontsize":15}
plt.pie(class1_students,labels=classes,explode=explode,colors=colores,
        autopct="%0.1f%%" ,radius=1.5 , textprops=textpreporty)             # To change font size of labels , % use textprops
plt.show()
```
