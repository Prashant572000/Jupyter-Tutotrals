<center><h1>Sub Plot</center></h1>

When we plot, so we plot in specific area, To divide this area in form of rows & cols, then plot multiple plots in rows & col is called subplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from matplotlib import style
```

```python
# Syntax- nrows=2, ncol=2, index(place)=1
plt.subplot(2,2,1)
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)                                # We can plot 4 subplots only , because we took 2RX2C
plt.subplot(2,2,4)                                # If we give 5 subplots , it gives error
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)                                
plt.subplot(2,2,4)                                
plt.show()                                           # To hide above information
```

# Draw Pie plot in Place of 4 Subplots

```python
plt.subplot(2,2,1)
plt.pie([1])                                 # Took any data, business data
plt.subplot(2,2,2)
plt.pie([1,2])
plt.subplot(2,2,3)
plt.pie([1,2,3])
plt.subplot(2,2,4)
plt.pie([1,2,3,4])
plt.show()     
```

```python
plt.subplot(3,2,1)                                         # Took 3 rows, 2 columns , index=1
plt.subplot(3,2,2)
plt.subplot(3,2,3)
plt.subplot(3,2,4)
plt.subplot(3,2,5)
plt.subplot(3,2,6)
plt.show()
```

Method 1 - plt.subplot(3,2,1) <br>

```python
plt.figure(figsize=(16,9))                                 # Adjust size
plt.subplot(3,2,1)                                         # If we took 2 data in same index they will overlap 
plt.subplot(3,2,2)
plt.subplot(3,2,3)
plt.subplot(3,2,4)
plt.subplot(3,2,5)
plt.subplot(3,2,6)
plt.show()
```
