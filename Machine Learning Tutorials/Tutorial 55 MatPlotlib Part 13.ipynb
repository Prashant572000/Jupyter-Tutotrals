```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg             # Used to read , show image & perform some specific operations 
```

```python
img1 = mpimg.imread('C:\\Users\\prash\\OneDrive\\Pictures\\download.png')        # Read image in another folder
plt.figure(figsize=(16,9))
plt.axis("off")                                                                  # Hide axis
plt.imshow(img1)                                                                 # Show img using imshow fn
plt.show()              
```
