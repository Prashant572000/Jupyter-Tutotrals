Correlation - It is a statistical measure of the relationship between 2 variables (X,Y). <br>
It gives value from -1 to 1 (Float). <br>
-1 - Perfect negative Correlation (Ex- X increases then Y decreases) <br>
1  - Perfect positive Correlation (Ex- X increases then Y increases) <br>
0  - NO Correlation (Ex- X,Y increases No affect to each other) <br>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data1 = pd.read_csv('E:\Coding Tutorials\Jupyter Tutorial\Machine Learning Tutorials\Tutorialcsvdata\Tut64.csv')     # read csv
data2 = data1.drop(columns=['Country Code' , 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name') 
data2    
```

```python
data2.corr()                                                           # Find correlation b/w no.
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(data2.corr(), annot=True, linewidth=5)
ax.tick_params(size=10 , color='w' , labelsize=10 , labelcolor='k')                    # Give some common parameters  
plt.title("Heatmap of Who Will Survive" , fontsize=30)
plt.show()
```

# Import dataset from scikit learn library

```python
from sklearn.datasets import load_breast_cancer             # This data comes from sklearn github library
cancer_dataset = load_breast_cancer()                       # Assign  load_dataset_cancer obj to cancer_dataset class  
cancer_dataset
```

This dataset comes from github repository, It is form of Python dictionary(Key:values) <br>
malignment - Brust cancer (0)<br>
besign     - No cancer (1)<br>

Total 31 features we have with target


# Convert This Dataset to DataFrame format

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'] , cancer_dataset['target']] , 
                        columns = np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

Can we train ML algo. using 31 features - Yes we can <br>
but inference depends on accuracy , It happens , some features out of 31 doesn't relate to target feature <br>
If we not get good accuracy, we use Feature Engineering(Find Correlation of dataset and show on heatmap) Then see <br>
Any feature doesn't relate with target feature , then we drop that feature.

```python
plt.figure(figsize=(25,25))
ax = sns.heatmap(cancer_df.corr() , annot = True , linewidth=5)
ax.tick_params(size=10 , color='k' , labelsize = 10 , labelcolor = 'k')
plt.title()
ax.set(title = "Heatmap of Breast Cancer Patients" , 
       xlabel = "")
plt.show()
```

We took features only which maximum +ve and less -ve for training ML Model, left features we dropped(to increase accuracy) <br>
Which feature is under 0 , we dropped AND Which feature is under 1 , we took for training ML Model
