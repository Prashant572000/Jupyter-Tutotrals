# Create a blank image

```python
import numpy as np
import cv2
blank_image = np.zeros((512,512,3), np.uint8)
blank_image.fill(0)
```

```python
cv2.imshow("blank_image", blank_image)
#print("blank_image", blank_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# save an image

```python
cv2.imwrite('D:/computer vision/blank_image.jpg', blank_image)
```

# Read an Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('D:/computer vision/Picture.jpg')
print(np.shape(img))
plt.imshow(img)
plt.show()
```

# Resize images

```python
resize_img =cv2.resize(img,(1000,1500))
plt.imshow(resize_img)
plt.show()
```

# Resize by ratio

```python
w_ratio = 0.5
h_ratio = 0.5
new_img=cv2.resize(img,(0,0),img,w_ratio,h_ratio)
plt.imshow(new_img)
plt.show()
```

#  Flipping an Image

```python
# Horizontal flip of original image
H_flipped = cv2.flip(img, 1)
#cv2.imshow("Flipped Horizontally", H_flipped)
plt.imshow(H_flipped)
plt.show()
```

```python
# Vertical flip of original image
V_flipped = cv2.flip(img, 0)
plt.imshow(V_flipped)
plt.show()
```

```python
# flipping horizontally and vertically simultaneously
flipped_img = cv2.flip(img, -1)
plt.imshow(flipped_img)
plt.show()
```
