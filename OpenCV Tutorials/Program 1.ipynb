```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('C:\\Users\\prash\\OneDrive\\Pictures\\p1.jpg')
print(np.shape(img))
plt.imshow(img)
plt.show()
```

```python
# Window name in which image is displayed
window_name = 'Image'
 
# Center coordinates
start_point =(250,170)
 
# Radius of circle
end_point = (40,4)
 
# Red color in BGR
color1 = (255, 0, 0)
color2 = (55, 45, 122)
 
# Line thickness of -1 px
font =cv2.FONT_HERSHEY_SIMPLEX
org=(130,150)
fontScale= 1
thickness = 2


img = cv2.rectangle(img, start_point, end_point, color1, thickness)
img = cv2.putText(img, 'Prashant' ,org,font,fontScale,color2,thickness)
```

```python
cv2.imshow(window_name, img)
plt.imgshow(img)
plt.show()
```
