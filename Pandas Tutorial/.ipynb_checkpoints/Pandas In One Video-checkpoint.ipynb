# Access Some Rows & Columns in Data

```python
newdf.loc[[1,2] , ['C' , 'D']]
```

```python
newdf.loc[: , ['C' ,'D']]                                           # : Means Show all Rows , But only specified Col.
```

```python
newdf.loc[[1,2] , :]                                                # Only specified Rows & Show all Cols 
```

# Run Query , Just like in SQL

```python
newdf.loc[(newdf['A']<0.3)]                                         # Gives Rows no. position Where 'A' <0.3
```

```python
newdf.loc[(newdf['A'] < 0.3) & (newdf['C'] > 0.2)]                      # Give val where A<0.3 , C>0.2
```

loc  - By access data elements their specified or defind Row(insex) or column name        use loc <br>
iloc - By access data elements numbers only, don't care what Row(index) & Column name is  use iloc

# How To Remove Duplicates

```python
df3 = pd.DataFrame({"Name" : ['Alfman' , 'Batsman' , 'Alfman'  ],
                   "Toy"  : [np.nan   , 'Batmobile', 'Bullwhip'],
                   "Born" : [pd.NaT , pd.Timestamp("1940-04-25") , pd.NaT]
                  })
```

```python
df3.head()
```

```python
#df3.drop_duplicates(subset=['Name'])                                # Drop 3rd Row
```

```python
#df3.drop_duplicates(subset=['Name'] , keep='first')     # keep='first' by default means Remove 1st Duplicate which is last row
```

```python
# df3.drop_duplicates(subset=['Name'] , keep='last')                  # Drop duplicates except last occurance
```

```python
df3.drop_duplicates(subset=['Name'] , keep=False)                   # Drop All Duplicates
```

```python
df3.head()
```

```python
df3.shape
```

```python
df3.info
```

```python
df3.info()
```

```python
df3.head()
```

```python
df3['Name'].value_counts(dropna=False)                   # dropna=False means not remove na
```

```python
df3['Toy'].value_counts(dropna=False)                    # dropna=False means not remove na
```

```python
df3['Toy'].value_counts(dropna=True)                     # dropna=True means remove na
```

```python
df3.head()
```

```python
df3.notnull()                                           # True where notnull else False
```

# Handle Excel Sheet using Pandas
