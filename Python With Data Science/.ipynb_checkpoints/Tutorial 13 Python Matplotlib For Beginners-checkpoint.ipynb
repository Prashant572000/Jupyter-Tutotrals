```python
%matplotlib inline
from  matplotlib import pyplot as plt
```

# Basic Line Plot chart

```python
plt.plot([1,2,3] , [2,5,7])
plt.show()
```

# Add Title and Labels

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3] , [2,5,7])
plt.title("Basic Data")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.show()
```

# Compare Muliple Data in Single Chart

```python
%matplotlib inline
from matplotlib import pyplot as plt
x1 = [1,5,4]
x2 = [2,10,8]
y1 = [3,7,3]
y2 = [5,8 , 9]
plt.plot(x1 , y1)
plt.plot(x2 , y2)

plt.show()
```

# Change Style of Plot

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,5,4]
x2 = [23,45,89]
y1 = [3,7,3]
y2 = [5,8 , 9]
plt.plot(x1 , y1)
plt.plot(x2 , y2)

plt.show()
```

# Give Name To Multiple Lables

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,5,4]
x2 = [23,45,89]
y1 = [3,7,3]
y2 = [5,8 , 9]
plt.plot(x1 , y1 , label='First')
plt.plot(x2 , y2 , label='Second')
plt.legend()
plt.show()
```

# Adjust Line Width 

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,5,4]
x2 = [23,45,89]
y1 = [3,7,3]
y2 = [5,8 , 9]
plt.plot(x1 , y1 , label='First' , linewidth = 5)
plt.plot(x2 , y2 , label='Second' ,linewidth = 3)
plt.legend()
plt.show()
plt.savefig("Tut13.png")
```

# To Plot Scatter

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,5,4]
x2 = [23,45,89]
y1 = [3,7,3]
y2 = [5,8 , 9]
plt.scatter(x1 , y1 , label='First' , linewidth = 5)
plt.plot(x2 , y2 , label='Second' , linewidth = 3)
plt.legend()
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [3,5,4]
x2 = [23,45,89]
y1 = [3,7,3]
y2 = [5,8 , 9]
plt.hist(x1 ,label='First' , linewidth = 15)
plt.plot(x2 , y2 , label='Second' , linewidth = 3)
plt.legend()
plt.show()
```
